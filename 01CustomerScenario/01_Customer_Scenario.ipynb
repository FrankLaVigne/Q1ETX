# 1. Customer Scenario Introduction

## 1.1 The Situation We're Walking Into

You are meeting a customer with the following reality:

- They have hundreds to thousands of documents
- They are PDFs, Word Docs, PowerPoint files, HTML, or anything really
- Many are scanned, poorly structured, or created over years or decades
- The documents contain domain-specific language, implicit rules, and institutional knowledge

The customer does *not* come in asking for:

- RAG
- Fine-tuning
- Synthetic data
- A specific framework

They come in asking something much simpler and harder:

**"We have all this information. How do we actually use AI with it?"**

## 1.2 What the Customer Has Likely Already Tried

Understanding what the customer has already attempted shapes how you position the conversation and sets realistic expectations on both sides.

Assume the customer has already:

- Uploaded a few documents into a chat interface
- Asked natural-language questions
- Seen *some* useful answers
- Seen *enough* wrong answers to lose trust

Common reactions you'll hear:

- "It kind of works..."
- "It answers some questions well..."
- "But we can't rely on it"
- "We can't scale this"
- "We can't explain why it answers the way it does"

> **Key insight: Interest exists, but confidence does not.**

## 1.3 Constraints That Actually Matter

Before any technical discussion, surface the constraints that shape the solution.

#### Data Constraints

- Documents are the system of record
- Some information is outdated but still referenced
- Structure is inconsistent across sources

#### Operational Constraints

- This must be **repeatable**
- This must survive **model changes**
- This must not rely on manual steps

#### Trust & Control Constraints

- Data may be sensitive or proprietary
- External cloud services may be restricted
- The customer must be able to explain results internally

> **Key insight:** This is not an "AI capability" problem. It's an **enterprise data quality and reliability** problem.

## 1.4 What Success Looks Like (From the Customer's Perspective)

The customer is *not* asking for perfect answers. They are asking for answers that are:

* **Grounded** clearly based on *their* documents

* **Consistent** the same question yields the same answer

* **Defensible:** they can explain *why* the system answered the way it did

If you pause here and ask: *"Would you trust this in front of your legal, compliance, or engineering team?"* -- you've framed the problem correctly.

## 1.5 Why We Start Where We Start

Make the sequencing explicit before the lab begins:

> "We are going to start by asking what the model can do *without* your data."

Not because that's the solution, but because:

- It gives us a **baseline**
- It makes improvement visible
- It prevents us from attributing gains to the wrong thing

> "You can't explain improvement unless you know what 'before' looked like."

## 1.6 How This Scenario Drives the Rest of the Lab

Every step in today's lab is a response to this scenario:

- If the model struggles, we don't panic
- If answers improve, we ask *why*
- If results plateau, we evaluate, not escalate blindly

> "Every step today exists to answer one question: *Is this good enough -- and how do we know?*"

**Next:** We establish the baseline -- what a model can (and can't) do *without* customer data.
